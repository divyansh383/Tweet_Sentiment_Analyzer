# Data Cleaning

```python
from nltk.corpus import stopwords
from string import punctuation
stop=stopwords.words('english')
stop += list(punctuation)
stop += ['flight','airline','flights','AA']
abbreviations = {'ppl': 'people','cust':'customer','serv':'service','mins':'minutes','hrs':'hours','svc': 'service',
           'u':'you','pls':'please','hr':'hour'}

def deEmojify(inputString):
    return inputString.encode('ascii','ignore').decode('ascii')
def hasNumbers(string):
    return any(char.isdigit() for char in string)
```

```python
#removing stop words
#removing mentions
#removing abbrev

import re


clean_text = []
for tweet,airline in train.loc[:, ["text","airline"]].values:
    
    tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','',tweet) #remove links
    tweet = re.sub('@[^\s]+','',tweet) #remove usernames
    tweet = re.sub('[\s]+', ' ', tweet) #remove additional whitespaces
    tweet = re.sub(r'#([^\s]+)', r'\1', tweet) #replace #word with word
    tweet = tweet.strip('\'"') #trim tweet   
    
    words=""  
    for word in tweet.split():
        if (word.lower() not in stop):
            word=deEmojify(word)
            if(hasNumbers(word)):
                print(word);
                continue;
            if word in list(abbreviations.keys()):
                words+=(abbreviations[word])+" "
            else:
                words+=(word.lower())+" "
    words+airline+" "            
    clean_text.append(words)
    
print(len(clean_text))
print(len(train.loc[:,'text']))
```

## Preparing Vocabulary

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
vector=TfidfVectorizer(analyzer='word',max_features=3000,max_df=0.8,ngram_range=(1,3))
train_features=vector.fit_transform(train.text)
test_features=vector.transform(test.text)
#vector.vocabulary_
train_features
```

```python
from sklearn.svm import SVC
clf=SVC(kernel="rbf",C=3);
clf.fit(train_features,train['airline_sentiment'])
clf
```

```python
predictions=clf.predict(train_features)
clf.score(train_features,train['airline_sentiment'])
```
